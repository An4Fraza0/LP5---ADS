# Usando o pandas

```python
frutas2 = ['laranja', 'manga', 'melancia']

frutas3 = pd.Series(frutas2)
print(type(frutas2))
print(type(frutas3))
```

```python
print(frutas2)
print(frutas3)
```

```python
# é o mesmo que print, porém usado no pandas
display(frutas3)
```

```python
# (df = dataframe) ao invés de nomear a variável como frutas4, nomeia-se como df, para facilitar a leitura do código, e também para seguir a convenção de nomeação do pandas.
frutas4 = pd.DataFrame({
    
    'nomes': ['laranja', 'manga', 'melancia'],
    'quantidade': [5, 2, 1],
    'valor': [3.00, 5.00, 10.00],
    'data': ['2026/04/01', '2026/04/02', '2026/04/03']

})

display(frutas4)
```
